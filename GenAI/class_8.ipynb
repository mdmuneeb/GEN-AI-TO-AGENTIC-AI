## Giving multiple conversation as a history to the model

```python
from langchain.messages import HumanMessage, AIMessage, SystemMessage 
from langchain_groq import ChatGroq
from dotenv import load_dotenv

load_dotenv()
llm = ChatGroq(
    model="llama-3.1-8b-instant", # qwen/qwen3-32b
    temperature=0,
)

conversation = [
    SystemMessage("You are a helpful assistant that translates English to French."),
    HumanMessage("Translate: I love programming."),
    AIMessage("J'adore la programmation."),
    HumanMessage("Translate: I love building applications.")
]

response = llm.invoke(conversation)
print (response.content)
# AIMessage("J'adore créer des applications.")
```

## Batch
- Batching a collection of independent requests to a
model can significantly improve performance and
reduce costs, as the processing can be done in
parallel

```python
from langchain_groq import ChatGroq
from dotenv import load_dotenv

load_dotenv()

llm = ChatGroq(
    model="llama-3.1-8b-instant",
    temperature=0)
l = [
    "What is the capital of France?",
    "What is the capital of Germany?",
    "What is the capital of Italy?"
]

response = llm.batch(l)
print(response)
```

```python
for i, r in enumerate(response):
    print(f"Q: {l[i]}")
    print(f"A: {r.content}")
```

## Stream
- Most models can stream their output content while
it is being generated. By displaying output
progressively, streaming significantly improves user
experience, particularly for longer responses.
- Output will come in the token by token 

```python

from langchain_groq import ChatGroq 
from dotenv import load_dotenv

load_dotenv()

llm = ChatGroq (
    model="qwen/qwen3-32b", # llama-3.1-8b-instant
    temperature=0,
)
query= "write a 500 words essay on agentic AI" 
streaming = llm.stream(query)

# while True:
#     try:
#         chunk = next(streaming)
#         print (chunk.text, end="", flush=True)
#     except StopIteration:
#         break

for chunk in streaming:
    print(chunk.text, end="", flush=True)
```

## Prompt Designing
- prompt design is the process of creating a prompt
tailored to the specific task that the system is being
asked to perform.
## Prompt Engineering
- it is the process of creating a prompt that is designed to
improve a performance

#### 2 Types of the Prompt Templates in LangChain
- PromptTemplate: For Single Turn Conversation
- ChatPromptTemplate: For Multi-Turn Conversation

- single turn = 1 input + 1 output
- multi turn = multiple inputs + multiple outputs


```python

from langchain_core.prompts import PromptTemplate 
from langchain_openai import ChatOpenAI
from dotenv import load_dotenv
from prompt_toolkit import prompt

load_dotenv()

llm = ChatGroq (
    model="qwen/qwen3-32b", # llama-3.1-8b-instant
    temperature=0,
)
# Define template with placeholder
template = PromptTemplate(
    input_variables=["country"],
    template="How many cities are in {country}?"
)
# Fill in the template dynamically

while True:
     country = input("Enter a country name (or 'exit' to quit): ")
     if country.lower() == 'exit':
         break
     prompt_filled = template.format(country=country)
     response = llm.invoke(prompt_filled)
     print(response.content)
```
