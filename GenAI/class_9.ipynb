```python
# Exercise - Ticket Classification
from langchain_core.prompts import PromptTemplate

template = """
You are a customer support assistant.

Your task is to classify the support ticket into one of the following categories:

1. Billing
2. Technical Issue
3. Account Access
4. General Inquiry

Read the ticket carefully and return ONLY the category name.

Ticket:
{ticket}

Category:
"""

tickets = [
"I was charged twice for my monthly subscription. Please refund the extra payment.",
"I cannot log into my account even after resetting my password.",
"The mobile app keeps crashing whenever I try to upload a photo.",
"Can you tell me what features are included in the premium plan?",
"My credit card was billed after I cancelled my subscription.",
"The website loads very slowly and sometimes shows a server error.",
"I forgot my password and the reset link is not working.",
"Do you offer discounts for students?",
"My payment failed but the amount was deducted from my bank.",
"I am unable to update the software on my device."
]

model = ChatGroq(
   model="llama-3.1-8b-instant",
   temperature=0 
)

prompt = PromptTemplate(
   template=template,
   input_variables=["ticket"]
)

for ticket in tickets:
    formatted_prompt = prompt.format(ticket=ticket)
    response = model.invoke(formatted_prompt)
    print("Ticket:", ticket)
    print("Category:", response.content)
    print()

# formatted_prompt = prompt.format(ticket=ticket)
# response = model.batch(formatted_prompt)
# print("Ticket:", ticket)
# print("Category:", response.content)
# print()
```

```python
from langchain_core.messages import SystemMessage, HumanMessage, AIMessage 
from langchain_groq import ChatGroq
from dotenv import load_dotenv
from langchain_core.messages.utils import trim_messages, count_tokens_approximately

load_dotenv()

model = ChatGroq(
    model="llama-3.1-8b-instant",
    temperature=0 
)

messages=[
    SystemMessage (content='You are a helpful assistant short ans with in 20 words'), 
    # HumanMessage(content='Tell me about LangChain') 
]


# result = model.invoke (messages)

# messages.append (AIMessage(result.content)) 

# print (messages)

while True:
    user_input=input("Enter: ")
    if user_input != "exit":
        messages.append(HumanMessage(content=user_input))
        result = model.invoke(messages)
        messages.append(AIMessage(content=result.content))
        print(f"AI: {result.content}")
        messages = trim_messages(
            messages, strategy="last", token_counter=count_tokens_approximately, max_tokens=40, start_on="human", end_on="human"
        )
        print(messages)
    else:
        break

    
```

```python

from langchain_openai import ChatOpenAI
from dotenv import load_dotenv
from typing import TypedDict, Annotated
from dotenv import load_dotenv

load_dotenv()

model = ChatOpenAI(model="gpt-4o-mini", api_key=OPENAI_API_KEY)

class ChatResponse(TypedDict):
    intent: Annotated[str, "The detected user intent (e.g., 'product_inquiry', 'order_status', 'greeting')"]
    confidence: Annotated [float, "A confidence score between 0 and 1 representing how sure the model is about the intent"]
    answer: Annotated [str, "The chatbot's primary response to the user query"]
    follow_up_questions: Annotated [list[str], "Suggested follow-up questions to keep the conversation flowing (can be empty)"]

structured_model=model.with_structured_output(ChatResponse)
result = structured_model.invoke(
    """Hi there! I ordered a smartwatch last week, but I haven't received any update yet. 
    Can you tell me the status of my order?"""
)

print(result)
```

```python
from pydantic import BaseModel, Field
from typing import List

load_dotenv()
class ChatResponse (BaseModel):
    intent: str = Field(..., description="The detected user intent (e.g., 'product_inquiry', 'order_status', 'greeting')")
    confidence: float = Field(..., ge=0, le=1, description="A confidence score between 0 and 1 representing how sure the model is about the intent")
    answer: str = Field(..., description="The chatbot's primary response to the user query")
    follow_up_questions: List[str] = Field (default_factory=list, description="Suggested follow-up questions to keep the conversation flowing (can be empty)")

model = ChatOpenAI(model="gpt-4o-mini", api_key=OPENAI_API_KEY)

sample_response = {
    "intent": "order_status",
    "confidence": 0.92,
    "answer": "Your smartwatch order is being processed and will be shipped within 2 business days.", 
    "follow_up_questions": ["Would you like to track your package?", "Do you want to update your delivery address?"]
}

structured_model=model.with_structured_output(ChatResponse)
result = structured_model.invoke(
    """Hi there! I ordered a smartwatch last week, but I haven't received any update yet. 
    Can you tell me the status of my order?"""
)
```
