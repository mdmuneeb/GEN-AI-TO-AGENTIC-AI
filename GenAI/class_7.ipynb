## Fundamentals of GenAI
- Artificial intelligence is the capability of computational
systems to perform tasks typically associated with human
intelligence, such as learning, reasoning, problem-solving,
perception, and decision-making.

## Generative Artificial Intelligence (GenAI)
- Generative AI, sometimes called gen AI, is artificial
intelligence (AI) that can create original content such as
text, images, video, audio or software code in response to
a user’s prompt or request.

## Discriminative & Generative Models
- Discriminative models are Machine learning and Deep learning based models
- Generative models are that generate data similar to training data

## Pre-Trained Model
- A pretrained model is a machine learning model that has been previously trained on a large dataset.
## Fine-Tuned Model
- A pre-trained model that is further trained on a smaller, task-specific dataset

## Foundation Model
- A foundation model is a large-scale AI model trained on massive and diverse datasets that can be adapted to a wide range of downstream tasks through fine-tuning or prompting

## Introduction To LLM:
- A large language model (LLM) is a type of artificial

- A Vision-Language Model is an AI model that

## What is Multimodal LLM?
- A foundational language model that can

```python
from dotenv import load_dotenv
from openai import OpenAI

load_dotenv()
client = OpenAI()

response = client.responses.create(
    model="gpt-4o-mini",
    input="Write a short story about a robot learning to fight in 2 lines."
)
print(response)
print(response.output_text)
```

#### if model is LLM only if it is giving output in the form of string

```python
# LLM Model

from langchain_openai import OpenAI
from dotenv import load_dotenv

load_dotenv()
client = OpenAI(
    model="gpt-3.5-turbo-instruct"
)

response = client.invoke("What is the capital of Pakistan?")

print(response)
```

```python
from langchain_openai import ChatOpenAI
from dotenv import load_dotenv

load_dotenv()
client = ChatOpenAI(model="gpt-4o-mini")

response = client.invoke("What is the capital of Pakistan?")

print(response.content)
```

## Temperature
- Output generate tokens ma hi hota aur haman text converted hi milta hai.
- temperature (0-2, default=0.7) text ka randomness ko dekhta hai
  - 0 par same hi output milraha
  - 2 ka qareeb hoga to output different text mai dega (eg. script writing)

## Max Completions tokens
- bounds the model to give output upto the max completions

## Max Tokens
- It does NOT count:
  - Your prompt
  - Chat history

```python
from langchain_openai import ChatOpenAI
from dotenv import load_dotenv

load_dotenv()
# client = ChatOpenAI(model="gpt-4o-mini", temperature=0.7, max_completion_tokens=30)
client = ChatOpenAI(model="gpt-4o-mini", temperature=0.7, max_tokens=30)

response = client.invoke("What are your thoughts about the Pakistan politics?")

print(response.content)
```

## Init Chat Models
- Another way to import chat models
- it just need to be given model name any (Openai, Gemini etc) but there integration package should be installed

```python
from langchain.chat_models import init_chat_model
from dotenv import load_dotenv

load_dotenv()
client = init_chat_model(model="gpt-4o-mini")

response = client.invoke("What are your thoughts about the Pakistan politics?")

print(response.content)
```

## Hugging Face

```python
from langchain_huggingface import ChatHuggingFace, HuggingFaceEndpoint
from dotenv import load_dotenv

load_dotenv()

llm = HuggingFaceEndpoint(
    repo_id="mistralai/Mistral-7B-Instruct-v0.2"
)

model = ChatHuggingFace(llm=llm)
result = model.invoke("What are your thoughts about the Pakistan politics?")
print(result)
print(result.content)
```
