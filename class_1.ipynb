### How to change data type of any variable (Type casting)
- float(): any int to float 
- int(): any float to int, when converting to int it removes all the number after the decimal point (.)
- incase of Boolean True we convert into the integer it gives 1. if any other thing present like characters it will give error
- if 7.57 is written in string then we convert it into the integer it will give error 
- Python supports this conversion float("45") into the float

```python
height = 10
print(type(height))
height = float(height)
print(type(height))

print()

number = 5.7
print(type(number))
number = int(number)
print(type(number))
# it removes all the numbers after decimal point
print(number)

print()

print(int(True))
print(type(int(True)))

print(int("4.57")) # --> Will give error   
```

### Dynamic String
- string concatenation
- f-string method
- string format

```python
name = input("Enter your Name: ")
age = input("Enter you Age: ")
print("String Concatenation")
print("My name is " + name + " and my Age is " + age)

print()
print("F-String")
print(f"My name is {name} and my age is {age}")

print()
print("String Fomat")
print("My name is {1} and my age is {0}".format(name, age)) # 0, 1 are index position
```
