## Loop

#### break

```python
for i in range(1, 11):
    if i == 5:
        break
    print(i)
```

#### Continue

```python
for i in range(1, 11):
    if i == 5:
        continue
    print(i)
```

#### Pass

```python
for i in range(1, 11):
    pass
```

## While Loop
- else is used to check that loop has executed fully

```python
x = 0
while x < 5:
    print(x)
    x += 1
else:
    print("Loop finished without break.")
print()
x = 0
while x < 5:
    print(x)
    if x == 3:
        break
    x += 1
else:
    print("Loop finished without break.")
```

## Comprehension
- list
- set
- distionary
- efficient at memory level
- code ziada readble hoga
- list ma is wajah sa nhi hoga kiunka wih mutable hai

```python
# List comprehension
items = [item for item in range(11)]
print(items)

# set comprehension
sets = {item for item in range(11)}
print(sets)

# dict comprehension
dicts = {f"key {item}": f"Value {item+1}" for item in range(11)}
print(dicts)

iterable = [("a", 1), ("b", 2), ("c", 3)]
result = {key: value for key, value in iterable}
print(result)

dicts = {item: item**2 for item in range(1, 11) if item % 2 == 0}
print(dicts)
```

#### agar sirf if lagana hai to if for ka baad hoga aur agar else lagana hai to for loop sa pehla if else aiga

```python
list = [item for item in range(11) if item % 2 == 0]
print(list)

list = [item if item % 2 == 0 else "Not a even number" for item in range(11) ]
print(list)
```

#### Function with docstring
- function ka document

```python
def check_divisible_by_3(num):
    """
    In this we are checking the number is divisible by 3 or not and returning the result as string.
    """
    if num % 3 == 0:
        return "Divisible by 3"
    else:
        return "Not Divisible by 3"
    
print(check_divisible_by_3(10))
```

```python
def check_prime(num):
    """We are basically checking the number is prime or not and returning the result as string."""
    for i in range(2, num):
        if num % i == 0:
            return f"{num} is not a prime number."
        
    return f"{num} is a prime number."

print(check_prime(17))
```

## Practise Problem

```python
def check_even_odd(num):
    if num % 2 == 0:
        return True
    else:
        return False
    
list = [i for i in range(1, 11)]
listEven = [item for item in list if check_even_odd(item)]
listOdd = [item for item in list if not check_even_odd(item)]

print(listEven)
print(listOdd)
```
