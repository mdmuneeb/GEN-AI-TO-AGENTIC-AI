### Arithmetic Operators
- Addition
- Subtraction
- Multiplication
- Division (/) normal division
- Division (//) float division
- Modulus (Remainder)
- Power

```python
a = 2 + 3
print(a)

a = 2 - 3
print(a)

a = 2 * 3
print(a)

a = 2 / 3
print(a)

a = 2 // 3
print(a)

a = 20 % 3
print(a)

a = 2 ** 3
print(a)
```

### Logical Operators
- AND
- OR
- NOT

```python
print(1 and 0)

print(1 or 0)

print(not 1)
```

### Assignment Operators

```python
b = 2
b = b + 2
print(b)

# OR

b = 2
b +=2 # efficient way to write b = b + 2
print(b)

b = 2
b -=2 # efficient way to write b = b - 2
print(b)

b = 2
b *=2 # efficient way to write b = b * 2
print(b)

b = 5
b /= 2 # efficient way to write b = b / 2
print(b)

b = 5
b //= 2 # efficient way to write b = b // 2
print(b)

b = 5
b %= 2 # efficient way to write b = b % 2
print(b)

b = 5
b **= 2 # efficient way to write b = b ** 2
print(b)
```

### Practise Qusetion

```python
user_input_1 = int(input("Enter a reg fee: "))
user_input_2 = int(input("Enter a monthly fee: "))
user_input_3 = int(input("Enter number of months: "))

total_cost = (user_input_1 + (user_input_2 * user_input_3))

print(total_cost)

user_input_4 = input("Do you have any discount: ")

discount = "yes" == user_input_4

final_amount = total_cost - (total_cost * 0.1 * discount)

print(final_amount)
```

```python
from datetime import datetime, timedelta

print(datetime.now()) # it is in the fomat of date time
print(type(datetime.now()))
print()
# string from time
formated_date = datetime.now().strftime("Date = %d/%m/%y, Time = %H:%M:%S")
print(formated_date)
print(type(formated_date))
print()
# date from string
formated_date_rev = datetime.strptime(formated_date, "Date = %d/%m/%y, Time = %H:%M:%S")
print(formated_date_rev)
print(type(formated_date_rev))
print()
# Adding 5 days to current date
fomat_date_rev = datetime.now() + timedelta(days=5)
print(fomat_date_rev)
print(type(fomat_date_rev))
```

### Container Data Types
- sequential Data type
  - List: Sequential
    - Hetrogenous (different datatypes data can be stored in this)
    - execution is slow
    - occupy more space
- mapping data types

```python
Sports = ["Cricket", "Football", "Hockey"]
print(Sports[0])

print(Sports[0:])

print(Sports[0:2])

print(Sports[-1])

print(Sports[-2:])

print(Sports[-3:])

print(Sports[-3:-1])

print(Sports[-3:-2])
```
