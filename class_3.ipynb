#### Delete
- Del
- Pop

```python
L = [1,2,3,4,5,6,7,8]
del L[2:5]
print(L)
del L
print(L) # all list got deleted
```

```python
# Pop
L = [1,2,3,4,5,6,7,8,9]
L.pop()
print(L)
L.pop(2)
L
```

## Sets
- Unordered
- Mutable
- No Duplicates
- Cant contain mutable datatypes

```python
s = {}
print(s)
print(type(s))
s = set()
print(s)
print(type(s))
```

#### Automatically removes the duplicates

```python
s = {1,2,3,4,5,5, 6, 6, 6}
print(s)
```

#### Cant have mutable items

```python
s = {1,2,3,4,5, {6,7,8}}
print(s)
```

#### Cannot perform slicing or indexing because of unordered

#### Items can be added in this
- Add
- Update

#### delete 
- Del keyword
- Remove (if not found that particular item in the set then it will through error)
- Discard
- Pop

```python
s = set([1,2,3,4,5])
print(s)
s.remove(3)
print(s)
s.remove(100) # will give error if that item is not present in the set
print(s)
```

```python
s.discard(100) # will not give error if that item is not present in the set
print(s)
```

```python
s = {1,2,4,5,6, False, 45.9, (1,2,4,5)}
print(s)
```

## Dictionary
- Mutable
- keys cant be duplicated
- keys cant be mutable datatypes

```python
d = {}
print(d)
print(type(d))
```

```python
d = {"name": "Muneeb", "age": 25, "city": "Karachi"}
print(d)
```

#### keys cannot be mutable

```python
d = {[1,2,4]: "abc"}
print(d)
```

```python
print(d["name"])
print(d.keys())
print(d.values())
```

```python
d = dict([(1, "Muhammad"), (2, "Muneeb"), (3, "Zafar")])
print(d)
```

#### Use Case for duplicate keys

```python
d = {1: 100, 2: 200, 1: 300}
print(d)
```
